```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


df = pd.read_csv('presskardu_orders.csv')

# --- 1. REVENUE BY AREA ---
area_revenue = df.groupby('area')['price'].sum().sort_values(ascending=False)
print("💰 Revenue by Area:")
print(area_revenue)

# --- 2. MOST POPULAR SERVICES ---
service_count = df['service'].value_counts()
print("\n👕 Most Popular Services:")
print(service_count)

# --- 3. ORDER STATUS BREAKDOWN ---
status_count = df['status'].value_counts()
print("\n📦 Order Status:")
print(status_count)

# --- 4. MONTHLY REVENUE TREND ---
month_order = ['December','January','February','March','April','May']
monthly_revenue = df.groupby('month')['price'].sum().reindex(month_order)
print("\n📅 Monthly Revenue:")
print(monthly_revenue)

# --- 5. BUSIEST DAY ---
day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
day_count = df['day_of_week'].value_counts().reindex(day_order)
print("\n📆 Busiest Days:")
print(day_count)

# --- PLOTS ---
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('PressKardu Analytics Dashboard', fontsize=16, fontweight='bold')

# Plot 1 - Revenue by Area
area_revenue.plot(kind='bar', ax=axes[0,0], color='steelblue')
axes[0,0].set_title('Revenue by Area')
axes[0,0].set_xlabel('')
axes[0,0].tick_params(axis='x', rotation=45)

# Plot 2 - Popular Services
service_count.plot(kind='bar', ax=axes[0,1], color='coral')
axes[0,1].set_title('Most Popular Services')
axes[0,1].tick_params(axis='x', rotation=45)

# Plot 3 - Monthly Revenue
monthly_revenue.plot(kind='line', ax=axes[1,0], color='green', marker='o')
axes[1,0].set_title('Monthly Revenue Trend')
axes[1,0].tick_params(axis='x', rotation=45)

# Plot 4 - Busiest Days
day_count.plot(kind='bar', ax=axes[1,1], color='purple')
axes[1,1].set_title('Orders by Day of Week')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('presskardu_analysis.png', dpi=150)
plt.show()
print("✅ Analysis complete!")
```
